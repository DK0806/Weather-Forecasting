###Weather forecasting is the task of forecasting weather conditions for a given location and time. With the use of weather data and algorithms, it is possible to predict weather conditions for the next n number of days.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
data = pd.read_csv("/content/DailyDelhiClimateTrain.csv")
print(data.head())
```

```python
print(data.describe())
```

```python
# lets have look in mean temp in delhi over the years
fig = px.line(data , x = "date",y="meantemp" , title = "Mean Temp in delhi over years")
fig.show()
```

```python
# humidity in delhi over years
fig = px.line(data , x = "date" , y="humidity" , title = "humid in delhi over years")
fig.show()
```

```python
# wind speed in delhi over years
fig = px.line(data , x='date' , y='wind_speed' , title = "wind speed in delhi over years")
fig.show()
```

```python
# relationship between temp and humidity
fig = px.scatter(data_frame=data , x = "humidity" ,y="meantemp" , trendline="ols" , title="relationship between temp and humidity")
fig.show()
```

### Now let’s analyze the temperature change in Delhi over the years. For this task,

`I will first convert the data type of the date column into datetime. Then I will add two new columns in the dataset for year and month values.`

```python
data['date'] = pd.to_datetime(data['date'] , format = "%Y-%m-%d")
data['year'] = data['date'].dt.year
data["month"] = data['date'].dt.month
print(data.head())
```

```python
forecast_data = data.rename(columns = {"date": "ds",
                                       "meantemp": "y"})
print(forecast_data)
```

```python
from prophet import Prophet
from prophet.plot import plot_plotly, plot_components_plotly
model = Prophet()
model.fit(forecast_data)
forecasts = model.make_future_dataframe(periods=365)
predictions = model.predict(forecasts)
plot_plotly(model, predictions)
```
